```python
# Implementing the Simple Moving Average Strategy
# Attempting to mimic Pseudo HFT trading algorithm using Backtrader (using days instead of seconds, minutes)
# Swing Trading Strategy

class SmaCross(bt.Strategy):
    # list of parameters which are configurable for the strategy
    params = dict(
        pfast=1,  # period for the fast moving average
        pslow=3  # period for the slow moving average
    )

    def __init__(self):
        sma1 = bt.ind.SMA(period=self.p.pfast)  # fast moving average
        sma2 = bt.ind.SMA(period=self.p.pslow)  # slow moving average
        self.crossover = bt.ind.CrossOver(sma1, sma2)  # crossover signal

    def next(self):
        if not self.position:  # not in the market
            if self.crossover > 0:  # if fast crosses slow to the upside
                self.buy()  # enter long

        elif self.crossover < 0:  # in the market & cross to the downside
            self.close()  # close long position

cerebro = bt.Cerebro()  # create a "Cerebro" engine instance

# Set our desired cash start to $100,000
cerebro.broker.setcash(100000)
print('Starting Portfolio Value using Swing Trading Strategy for BTC: $%.2f' % cerebro.broker.getvalue())

# Creating a data feed
data1 = bt.feeds.YahooFinanceCSVData(dataname='Resources/BTC-USD.csv')
cerebro.adddata(data1)
cerebro.addstrategy(SmaCross)  # Add the trading strategy
cerebro.run()  # run it all

%matplotlib widget
%matplotlib inline
btc_swing_plot = cerebro.plot(iplot=False, volume=True, savefig=True, figfilename='backtrader-plot.png')  # and plot it with a single command

print('Final Portfolio Value using Swing Trading Strategy for BTC: $%.2f' % cerebro.broker.getvalue())
btc_swing_value = cerebro.broker.getvalue()
initial_investment = 100000
pnl_swing_btc = round(btc_swing_value - initial_investment, 2)
print(f'The Swing Trading Strategy for BTC had a ${pnl_swing_btc} profit over its initial investment of ${initial_investment}.')
```
